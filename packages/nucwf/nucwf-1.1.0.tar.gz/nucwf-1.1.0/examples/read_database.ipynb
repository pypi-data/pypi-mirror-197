```python
# load packages ....
%load_ext autoreload
%autoreload 2
import sys
sys.path.append("../src") 
from nucwf import access
# silent excessive warnings  
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
# set pandas columns to no limit 
import pandas as pd 
pd.set_option('display.max_columns', None)
```

```python
# connect to standard data base for wave functions 
wfdb=access.database(workdir="/Users/andreasnogga/work/wflocal")
```

```python
# show current wave functions available 
wfdb.pddf
```

```python
# cell retrieves wf number 4 from table above 
import os 
# get file info from table 
row=wfdb.pddf.loc[[4]]
wfid=row.to_dict('records')[0]
print(wfid)
# retrieve file
hashname,uniquename=wfdb.get_file(**wfid)
# print name and file size 
print("hashname:    ",hashname)
print("filesize:    ",os.path.getsize(hashname))
print("uniquename:  ",uniquename)
```

```python
# this is a A=4 file (see above)
wavein=access.wavefa4(hashname,printlevel=1)
```

```python
# print selected columns of table for easier choice of wave function 
slicedf=wfdb.pddf[(wfdb.pddf.lambdaSRGNN==0)][["Z","N","E[MeV]","MODENN","orderNN","LambdaNN","potnr","tnforder","cE1","lambdaSRGNN","benchnr"]].sort_values(by=["Z","N","tnforder","orderNN","LambdaNN","benchnr","lambdaSRGNN"])

print(slicedf.to_markdown(index=True))
```

```python
# download all wf with a specific benchnr and check basis properties 
# cell retrieves wf number 4 from table above 
import os 
# get wf with benchnr=22
select=(wfdb.pddf.benchnr>=22)
subdf=wfdb.pddf[select]

for index,row in subdf.iterrows():
   rowval=wfdb.pddf.loc[[index]]
   wfid=rowval.to_dict('records')[0]
   # retrieve file
   hashname,uniquename=wfdb.get_file(**wfid)
   # print name and file size 
   print("hashname:    ",hashname)
   print("filesize:    ",os.path.getsize(hashname))
   print("uniquename:  ",uniquename)
   # this is A=4, so read 4He wf  
   wavein=access.wavefa4(hashname,printlevel=1)
```
